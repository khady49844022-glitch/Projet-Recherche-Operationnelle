```python
# ====================================
# PROJET RO 2026
# TSP - Voyageur de Commerce
# ====================================

import itertools

villes = ["A", "B", "C", "D"]

distances = {
    ("A", "B"): 10,
    ("A", "C"): 15,
    ("A", "D"): 20,
    ("B", "C"): 35,
    ("B", "D"): 25,
    ("C", "D"): 30
}

def distance(a, b):

    if (a, b) in distances:
        return distances[(a, b)]

    return distances[(b, a)]

min_route = None
min_distance = float("inf")

for permutation in itertools.permutations(villes):

    total = 0

    for i in range(len(permutation) - 1):

        total += distance(
            permutation[i],
            permutation[i+1]
        )

    total += distance(
        permutation[-1],
        permutation[0]
    )

    if total < min_distance:

        min_distance = total
        min_route = permutation

print("Meilleure route :", min_route)
print("Distance minimale :", min_distance)
```
