```python
# =========================
# PROJET RO 2026
# PROBLEME SAC A DOS
# =========================

# Données du problème

objets = ["Objet A", "Objet B", "Objet C"]

poids = [2, 3, 4]

valeurs = [3, 4, 5]

capacite = 5

n = len(valeurs)

# Création tableau dynamique

dp = [[0 for x in range(capacite + 1)] for x in range(n + 1)]

# Algorithme programmation dynamique

for i in range(n + 1):
    
    for w in range(capacite + 1):
        
        if i == 0 or w == 0:
            dp[i][w] = 0

        elif poids[i-1] <= w:

            dp[i][w] = max(
                valeurs[i-1] + dp[i-1][w-poids[i-1]],
                dp[i-1][w]
            )

        else:
            dp[i][w] = dp[i-1][w]

# Résultat final

print("Valeur maximale =", dp[n][capacite])
```
